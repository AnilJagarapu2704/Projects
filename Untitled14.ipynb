### **1. How has the number of movies released per year changed over the last 20-30 years?**

The line plot here suggests that there is a keen spike in the movies releases over the years from starting at very low near to 20 movies per year to huge raise in recent years. There is dip in the movie releases from past 1-2 years but as per the trend it is likely to increase again.

###**3. What is the best time to launch a TV show?**

##Non-Graphical Analysis

```python
# Number of shows released on each day

data = data.copy()
valid_data = data.drop_duplicates(subset = 'show_id')
shows = valid_data[valid_data['type'] == 'TV Show']
valid_dates = shows[shows['date_added'] != 'Unknown']

valid_dates['date_added'] = valid_dates['date_added'].str.strip()
pd.options.mode.chained_assignment = None
valid_dates['date_added'] = pd.to_datetime(valid_dates['date_added'])

valid_dates['day'] = valid_dates['date_added'].dt.day
valid_dates['week'] = valid_dates['date_added'].dt.day_name()
valid_dates['month'] = valid_dates['date_added'].dt.month_name()
valid_dates['year'] = valid_dates['date_added'].dt.year

daily_show_count = valid_dates.groupby('day')['show_id'].count().reset_index()
daily_show_count.rename(columns = {'show_id': "show_count"}, inplace = True)
daily_show_count
```

```python
# Number of shows released on weekday

weekly_count = valid_dates.groupby('week')['show_id'].count().reset_index()
weekly_count.rename(columns = {'week': 'week_day', 'show_id': 'show_count'}, inplace = True)

week_day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekly_count['week_day'] = pd.Categorical(weekly_count['week_day'], categories = week_day_order, ordered = True)
weekday_count = weekly_count.sort_values('week_day')
weekday_count
```

```python
# Number of shows released on each month

monthy_count = valid_dates.groupby('month')['show_id'].count().reset_index()
monthy_count.rename(columns = {'show_id': 'show_count'}, inplace = True)

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthy_count['month'] = pd.Categorical(monthy_count['month'], categories = month_order, ordered = True)
ordered_month_count = monthy_count.sort_values('month')
ordered_month_count
```

```python
# Number of shows released per year

yearly_count = valid_dates.groupby('year')['show_id'].count().reset_index()
yearly_count.rename(columns = {'show_id': 'show_count'}, inplace = True)
yearly_count
```

##Visual Analysis

```python
plt.figure(figsize = (15, 9))
plt.subplot(2, 1, 1)
sns.barplot(data = daily_show_count, x = 'day', y = 'show_count', color = 'steelblue')
plt.title('Number of shows per day')
plt.xlabel('Day', size = 7)
plt.ylabel('Show Count')
plt.subplot(2, 3, 4)
pie_values = weekday_count['show_count']
week_day = weekday_count['week_day']
plt.pie(pie_values, labels = week_day, autopct = lambda x: f"{int(x * sum(pie_values)/100)}", startangle = 90)
plt.title('Number of shows per each day of week')
plt.subplot(2, 3, 5)
sns.barplot(data = ordered_month_count, x = 'month', y = 'show_count', color = 'orange')
plt.title('Number of shows per month')
plt.xlabel('Month')
plt.ylabel('Show Count')
plt.xticks(rotation = 90)
plt.subplot(2, 3, 6)
sns.lineplot(data = yearly_count, x = 'year', y = 'show_count', color = 'green', marker = 'o', linewidth = 2)
plt.title('Number of shows per year')
plt.xlabel('Year')
plt.ylabel('Show Count')
plt.suptitle('Number of shows over various parts of year')
plt.show()
```

The above plot describes that start, mid, end days in a month are when more TV shows got uploaded for viewership. Mostly TV shows are targeted to be released on Friday might be because of friday being the start of the weekend. During second half of the year (July - December) more TV shows are released compared to first half of the year. Since 2015 the volume of TV shows increased in Netflix.

```python
heat_data = valid_dates.groupby(['month', 'year'])['show_id'].count().unstack()
sns.heatmap(heat_data, cmap='Spectral_r')
plt.title('TV Show Release Heatmap (Month vs Weekday)')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

Above Heatmap clearly illustrates that from 2008 to 2014 there are none to bare minimum TV shows. This changed from 2015 signficantly by shows getting released almost every month of the year, sometimes the number getting to upwards of 50.

###**5. Does Netflix has more focus on TV Shows than movies in recent years**

##Non-Graphical Analysis

```python
# Comparison of movie count vs TV show count over past decade

valid_data = data.drop_duplicates(subset = 'show_id')
valid_dates = valid_data[valid_data['date_added'] != 'Unknown']

valid_dates['date_added'] = valid_dates['date_added'].astype(str).str.strip()
valid_dates['date_added'] = pd.to_datetime(valid_dates['date_added'])

valid_dates['day'] = valid_dates['date_added'].dt.day
valid_dates['month'] = valid_dates['date_added'].dt.month_name()
valid_dates['year'] = valid_dates['date_added'].dt.year

latest_year = valid_dates['year'].max()
past_decade = latest_year - 9
recent_years = valid_dates[valid_dates['year'] >= past_decade]
recent_years = recent_years.groupby(['year', 'type'], observed = False)['show_id'].count().reset_index()
recent_years.rename(columns = {'show_id': 'show_count'}, inplace = True)
recent_years
```

```python
# Growth Rate Analysis

recent_years = valid_dates[valid_dates['year'] >= past_decade]
growth_analysis = recent_years.groupby(['year', 'type'], observed = False)['show_id'].count().reset_index()

growth_analysis['growth_rate'] = growth_analysis.groupby('type', observed = False)['show_id'].pct_change() * 100
growth_analysis['growth_rate'] = growth_analysis['growth_rate'].replace([np.inf, -np.inf], np.nan).round(2)
growth_analysis['growth_rate'] = growth_analysis['growth_rate'].fillna(0)
growth_analysis
```

##Visual Analysis

```python
plt.figure(figsize = (10, 5))
sns.barplot(data = recent_years, x = 'year', y = 'show_count', hue = 'type')
plt.title('Comparison of Movies vs TV Shows over past decade')
plt.xlabel('Year')
plt.ylabel('Show Count')
plt.show()
```

The above plot depicts that over past decade from 2012 to 2021 every year number of movies added in Netflix are higher compared to TV shows. This trend eventually spiked from 2017 that number of movies uploaded are almost twice the amount with respect to TV shows.

```python
plt.figure(figsize = (10, 6))
sns.lineplot(data = growth_analysis, x = 'year', y = 'growth_rate', hue = 'type', marker = 'o', linewidth = 2)
plt.title('Growth Rate Analysis of each type')
plt.xlabel('Year')
plt.ylabel('Growth Percentage from prev year')
plt.show()
```

The above plot follows a cyclic trend. It illustrates that growth rate of TV shows is high compared to movies. The trend went upward untill 2017, but since 2017 the trend is downward trend. It went below zero from 2020 ownwards. The key here is that growth rate of TV shows is higher than movies. This suggests that Netflix is focusing on TV shows more.

```python
# Top 10 countries based on genre

valid_data = data.drop_duplicates(subset = 'show_id')
valid_countries = valid_data[valid_data['country'] != 'Unknown']

country_wise = valid_countries.groupby('country')['show_id'].count().reset_index()
country_wise.rename(columns = {'show_id': 'total_count'}, inplace = True)

country_wise.sort_values('total_count', ascending = False, inplace = True)
top_10_countries = country_wise.iloc[:10]
top_10_countries
```

```python
# Top 5 genre's

genre_sort = data.groupby('listed_in')['show_id'].count().reset_index().sort_values('show_id', ascending = False)
genre_sort.rename(columns = {'show_id': 'genre_count'}, inplace = True)
top_genres = genre_sort.iloc[:5]
top_genres
```

```python
# Performance of top genre's within top 10 countries

country_wise_genre_count = data.groupby(['country', 'listed_in'])['show_id'].count().reset_index()
country_wise_genre_count.rename(columns = {'show_id': 'genre_count'}, inplace = True)
country_wise_genre_count.sort_values('genre_count', ascending = False)

top_10_country_genres = country_wise_genre_count[country_wise_genre_count['country'].isin(top_10_countries['country'])]
top_genres_among_top_countries = top_10_country_genres[top_10_country_genres['listed_in'].isin(top_genres['listed_in'])]
top_genres_among_top_countries
```

##Visual Analysis

```python
plt.figure(figsize = (15, 6))
sns.barplot(data = top_genres_among_top_countries, x = 'country', y = 'genre_count', hue = 'listed_in')
plt.title('Performance of top genres in top 10 countries')
plt.xlabel('Country')
plt.ylabel('Genre Count')
plt.legend(title = 'Genre')
plt.show()
```

The above barplot illustrates that International movies/ TV shows are popular among all top countries except United States and Canada. Western countries like United States, United Kingdom and  Canada are heavily intreseted in Comedies.
