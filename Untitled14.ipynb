### **1. How has the number of movies released per year changed over the last 20-30 years?**

## Non-Graphical Analysis

```python
curr_year = data['release_year'].max()
prev_year = curr_year - 29
required_years = data[(data['release_year'] >= prev_year) & (data['release_year'] <= curr_year) & (data['type'] == 'Movie')]

tidy_data = required_years.drop_duplicates(subset = 'show_id')
movies_per_year = tidy_data.groupby('release_year')['show_id'].count().reset_index(name = 'movie_count')
movies_per_year.sort_values('release_year', inplace = True)
movies_per_year
```

##Visual Analysis

```python
plt.figure(figsize = (12, 5))
x = movies_per_year['release_year']
y = movies_per_year['movie_count']
plt.plot(x, y)
plt.title(f'Movies over Years ({prev_year} - {curr_year})')
plt.xlabel('Year')
plt.ylabel('Movies Count')
plt.show()
```

The line plot here suggests that there is a keen spike in the movies releases over the years from starting at very low near to 20 movies per year to huge raise in recent years. There is dip in the movie releases from past 1-2 years but as per the trend it is likely to increase again.

```python
plt.figure(figsize = (15, 8))
sns.barplot(data = movies_per_year, x = 'release_year', y = 'movie_count')
plt.title(f'Movies over Years ({prev_year} - {curr_year})')
plt.xlabel('Year')
plt.ylabel('Movies Count')
plt.show()
```

This shows how releases are distributed over the years. We can see that over last 30 years the bars gradually tend to increase in size and peaked from 2016 onwards. This suggests Netflix’s expansion strategy intensified from recent years.

###**3. What is the best time to launch a TV show?**

##Non-Graphical Analysis

```python
# Number of shows released on each day

valid_data = data.drop_duplicates(subset = 'show_id')
shows = valid_data[valid_data['type'] == 'TV Show']
valid_dates = shows[shows['date_added'] != 'Unknown']

valid_dates['date_added'] = valid_dates['date_added'].str.strip()
pd.options.mode.chained_assignment = None
valid_dates['date_added'] = pd.to_datetime(valid_dates['date_added'])
valid_dates['day'] = valid_dates['date_added'].dt.day
valid_dates['week'] = valid_dates['date_added'].dt.day_name()
valid_dates['month'] = valid_dates['date_added'].dt.month_name()
valid_dates['year'] = valid_dates['date_added'].dt.year

daily_show_count = valid_dates.groupby('day')['show_id'].count().reset_index()
daily_show_count.rename(columns = {'show_id': "show_count"}, inplace = True)
daily_show_count
```

```python
# Number of shows released on weekday

weekly_count = valid_dates.groupby('week')['show_id'].count().reset_index()
weekly_count.rename(columns = {'week': 'week_day', 'show_id': 'show_count'}, inplace = True)
week_day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekly_count['week_day'] = pd.Categorical(weekly_count['week_day'], categories = week_day_order, ordered = True)
weekday_count = weekly_count.sort_values('week_day')
weekday_count
```

```python
# Number of shows released on each month

monthy_count = valid_dates.groupby('month')['show_id'].count().reset_index()
monthy_count.rename(columns = {'show_id': 'show_count'}, inplace = True)

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthy_count['month'] = pd.Categorical(monthy_count['month'], categories = month_order, ordered = True)
ordered_month_count = monthy_count.sort_values('month')
ordered_month_count
```

```python
# Number of shows released per year

yearly_count = valid_dates.groupby('year')['show_id'].count().reset_index()
yearly_count.rename(columns = {'show_id': 'show_count'}, inplace = True)
yearly_count
```

##Visual Analysis

```python
plt.figure(figsize = (15, 9))
plt.subplot(2, 1, 1)
sns.barplot(data = daily_show_count, x = 'day', y = 'show_count', color = 'steelblue')
plt.title('Number of shows per day')
plt.xlabel('Day', size = 7)
plt.ylabel('Show Count')
plt.subplot(2, 3, 4)
pie_values = weekday_count['show_count']
week_day = weekday_count['week_day']
plt.pie(pie_values, labels = week_day, autopct = lambda x: f"{int(x * sum(pie_values)/100)}", startangle = 90)
plt.title('Number of shows per each day of week')
plt.subplot(2, 3, 5)
sns.barplot(data = ordered_month_count, x = 'month', y = 'show_count', color = 'orange')
plt.title('Number of shows per month')
plt.xlabel('Month')
plt.ylabel('Show Count')
plt.xticks(rotation = 90)
plt.subplot(2, 3, 6)
sns.lineplot(data = yearly_count, x = 'year', y = 'show_count', color = 'green')
plt.title('Number of shows per year')
plt.xlabel('Year')
plt.ylabel('Show Count')
plt.suptitle('Number of shows over various parts of year')
plt.show()
```

The above plot describes that start, mid, end days in a month are when more TV shows got uploaded for viewership. Mostly TV shows are targeted to be released on Friday might be because of friday being the start of the weekend. During second half of the year (July - December) more TV shows are released compared to first half of the year. Since 2015 the volume of TV shows increased in Netflix.

```python
heat_data = valid_dates.groupby(['month', 'year'])['show_id'].count().unstack()
sns.heatmap(heat_data, cmap='Spectral_r')
plt.title('TV Show Release Heatmap (Month vs Weekday)')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

Above Heatmap clearly illustrates that from 2008 to 2014 there are none to bare minimum TV shows. This changed from 2015 signficantly by shows getting released almost every month of the year, sometimes the number getting to upwards of 50.

###**4. Analysis of actors/directors of different types of shows/movies.**

##Non-Graphical Analysis

```python
# Top 10 directors overall

data['cast'] = data['cast'].astype(str).str.strip()
data['director'] = data['director'].astype(str).str.strip()

tidy_data = data.drop_duplicates(subset='show_id')
directors = tidy_data[(tidy_data['director'] != '') & (tidy_data['director'] != 'Unknown')]

director_count = directors['director'].value_counts().reset_index()
director_count.rename(columns = {'count': 'appearence_count'}, inplace = True)
top_10_directors_overall = director_count.iloc[:10]
top_10_directors_overall
```

```python
# Top 10 actors overall

tidy_data = data.drop_duplicates(subset='show_id')
actors = tidy_data[(tidy_data['cast'] != 'Unknown') & (tidy_data['cast'] != '')]

actor_count = actors['cast'].value_counts().reset_index()
actor_count.rename(columns = {'count': 'appearence_count'}, inplace = True)
top_10_actors_overall = actor_count.iloc[:10]
top_10_actors_overall
```

```python
# Top actors in Movies

tidy_data = data.drop_duplicates(subset='show_id')
only_movies = tidy_data[tidy_data['type'] == 'Movie']

mov_actors = only_movies[(only_movies['cast'] != 'Unknown') & (only_movies['cast'] != '')]
movie_actor_count = mov_actors['cast'].value_counts().reset_index()
movie_actor_count.rename(columns = {'count': 'movie_count'}, inplace = True)
top_10_movie_actors = movie_actor_count.iloc[:10]
top_10_movie_actors
```

```python
# Top actors in TV shows

tidy_data = data.drop_duplicates(subset='show_id')
only_tvshows = tidy_data[tidy_data['type'] == 'TV Show']

tvshow_actors = only_tvshows[(only_tvshows['cast'] != 'Unknown') & (only_tvshows['cast'] != '')]
tvshow_actor_count = tvshow_actors['cast'].value_counts().reset_index()
tvshow_actor_count.rename(columns = {'count': 'show_count'}, inplace = True)
top_10_tvshow_actors = tvshow_actor_count.iloc[:10]
top_10_tvshow_actors
```

```python
# Top directors in Movies

tidy_data = data.drop_duplicates(subset='show_id')
only_movies = tidy_data[tidy_data['type'] == 'Movie']

mov_directors = only_movies[(only_movies['director'] != 'Unknown') & (only_movies['director'] != '')]
movie_director_count = mov_directors['director'].value_counts().reset_index()
movie_director_count.rename(columns = {'count': 'movie_count'}, inplace = True)
top_10_movie_directors = movie_director_count.iloc[:10]
top_10_movie_directors
```

```python
# Top directors in TV shows

tidy_data = data.drop_duplicates(subset='show_id')
only_tvshows = tidy_data[tidy_data['type'] == 'TV Show']

tvshow_directors = only_tvshows[(only_tvshows['director'] != 'Unknown') & (only_tvshows['director'] != '')]
tvshow_director_count = tvshow_directors['director'].value_counts().reset_index()
tvshow_director_count.rename(columns = {'count': 'show_count'}, inplace = True)
top_10_tvshow_directors = tvshow_director_count.iloc[:10]
top_10_tvshow_directors
```

```python
# Top Actor-Director pairs

data['director'] = data['director'].astype(str).str.strip()
data['cast'] = data['cast'].astype(str).str.strip()
pairs = data[((data['cast'] != 'Unknown') & (data['director'] != 'Unknown'))]
pairs = pairs.drop_duplicates(subset = ['show_id', 'director', 'cast'])
actor_director_pairs = pairs.groupby(['director', 'cast'])['show_id'].count().reset_index(name = 'count')
actor_director_pairs = actor_director_pairs.sort_values('count', ascending = False)
top_pairs = actor_director_pairs.iloc[:20]
top_pairs
```

##Visual Analysis

```python
plt.figure(figsize = (13, 6))
sns.barplot(data = top_pairs, x = 'cast', y = 'count', hue = 'director')
plt.title('Top actor-director pairs')
plt.xlabel('Actor')
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```

The above plot is a visual represenation of top 20 actor-director pairs. 'Rajiv Chilaka' is the director who collabrated most with actors 'Rajesh Kawa' and 'Julie Tejwani'. There are different directors collabrated with diffrent actors like 'S.S. Rajamouli' collabrated with 'Anushka Shetty', 'Prabhas', 'Tamannaah Bhatia', 'Ramya Krishnan',etc. In the x-axis we can see the list of actors and the colour of each bar depicts the director. Each colour indicates a different director.

```python
plt.figure(figsize = (15, 8))
plt.subplot(3, 3, 1)
sns.barplot(data = top_10_movie_directors, x = 'director', y = 'movie_count')
plt.title('Top directors in Movies')
plt.xlabel('Director')
plt.ylabel('Movies Count')
plt.xticks(rotation = 90)
plt.subplot(3, 3, 2)
sns.barplot(data = top_10_directors_overall, x = 'director', y = 'appearence_count', color = 'Red')
plt.title('Top directors Overall')
plt.xlabel('Director')
plt.ylabel('Overall Count')
plt.xticks(rotation = 90)
plt.subplot(3, 3, 3)
sns.barplot(data = top_10_tvshow_directors, x = 'director', y = 'show_count', color = 'yellow')
plt.title('Top directors in TV Shows')
plt.xlabel('Director')
plt.ylabel('Shows Count')
plt.xticks(rotation = 90)
plt.subplot(3, 3, 7)
sns.barplot(data = top_10_movie_actors, x = 'cast', y = 'movie_count', color = 'orange')
plt.title('Top actors in Movies')
plt.xlabel('Actor')
plt.ylabel('Movies Count')
plt.xticks(rotation = 90)
plt.subplot(3, 3, 8)
sns.barplot(data = top_10_actors_overall, x = 'cast', y = 'appearence_count', color = 'green')
plt.title('Top actors Overall')
plt.xlabel('Actor')
plt.ylabel('Overall Count')
plt.xticks(rotation = 90)
plt.subplot(3, 3, 9)
sns.barplot(data = top_10_tvshow_actors, x = 'cast', y = 'show_count', color = 'pink')
plt.title('Top actors in TV Shows')
plt.xlabel('Actor')
plt.ylabel('Shows Count')
plt.xticks(rotation = 90)
plt.suptitle('Top actors & directors over different types')
plt.show()
```
